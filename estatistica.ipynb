```python
plt.figure(figsize=(20, 12))
plt.xticks(rotation=90)
barplot = sns.barplot(data=dfA['AP'].reset_index(),x='index', y='AP', palette="Blues_d")
barplot.set_title('Pressão Plantar Antepé(Grupo A)')
barplot.set_ylabel('Pressão Antepé')
barplot.set_xlabel('Indivíduo')
```

```python
plt.figure(figsize=(20, 12))
plt.xticks(rotation=90)
barplot = sns.barplot(data=dfB['AP'].reset_index(),x='index', y='AP', palette="Blues_d")
barplot.set_title('Pressão Plantar Antepé(Grupo B)')
barplot.set_ylabel('Pressão Antepé')
barplot.set_xlabel('Indivíduo')
```

```python
print('De acordo com os Gráficos podemos analisar que as pressões plantares\nde antepé do segundo grupo supera as do primeiro grupo.')
print('A média de pressão plantar no antepé no primeiro grupo é '+str(round(dfA['AP'].mean(),2))+'\nenquanto que a do segundo é '+str(round(dfB['AP'].mean(),2))+' .')
```
